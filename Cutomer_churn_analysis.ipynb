```python
df["TotalCharges"] =  df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float64")
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

## Converted 0 and 1 values  of SeniorCitizens  into "yes/no" values to make it easier to understand

```python
ax = sns.countplot(x = df['Churn'], data = df )
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb = df.groupby('Churn').agg({'Churn':"count"})
gb
plt.pie(gb['Churn'],labels= gb.index, autopct= "%1.2f%%")
plt.title("% of Churned Customers",fontsize = 10)
plt.show()
```

```python
plt.figure(figsize=(3,4))
sns.countplot(x = "gender", data=df,hue = 'Churn')
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(3,4))
sns.countplot(x = "SeniorCitizen", data=df,hue = 'Churn')
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()
percentage_data = count_data.div(count_data.sum(axis=1), axis=0) * 100

# Plotting
plt.figure(figsize=(4, 4))
percentage_data.plot(kind='bar', stacked=True)

# Adding titles and labels
plt.title("Churn by SeniorCitizen")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.xticks(rotation=0)

# Adding percentage labels
for i in range(percentage_data.shape[0]):
    for j in range(percentage_data.shape[1]):
        plt.text(i, percentage_data.values[i, j] / 2 + percentage_data.iloc[i, :j].sum(), 
                 f"{percentage_data.values[i, j]:.1f}%", 
                 ha='center', va='center', color='white')

plt.legend(title='Churn', bbox_to_anchor = (0.9,0.9)) # customized legend location 
plt.show()
```

```python
# comparative is greater percentage of people in Senior Citizen category have churned
```

```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set the number of rows and columns for subplots
n_cols = 3  # Number of columns in the subplot
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate the number of rows needed

# Create a figure with subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))

# Flatten axes array for easy iteration
axes = axes.flatten()

# Loop through each column and create a countplot
for i, column in enumerate(columns):
    sns.countplot(x=column, data=df, ax=axes[i],hue = df['Churn'])
    axes[i].set_title(f'Count of {column}')
    axes[i].set_ylabel('Count')
    axes[i].set_xlabel(column)

# Hide any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
